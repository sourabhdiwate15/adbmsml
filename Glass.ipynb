```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import ConfusionMatrixDisplay as cmd
from sklearn.model_selection import train_test_split as tts
from sklearn.tree import DecisionTreeClassifier as dtc
```

```python
kk = pd.read_csv("glass.csv")
```

```python
x = kk[['RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe']]
y = kk['Type']
```

```python
xt,xte,yt,yte = tts(x,y,test_size=0.2,random_state=12)
xt.shape,xte.shape,yt.shape,yte.shape
```

```python
model = dtc(criterion = 'entropy',random_state =0,max_depth = 300)
```

```python
model.fit(xt,yt)
```

```python
model.score(xte,yte)
```

```python
prediction = model.predict(xte)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(yte,prediction)
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(15,15))
plot_tree(model,max_depth=20);
```
