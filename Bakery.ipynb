```python
import pandas as pd
```

```python
df=pd.read_csv('Bakery.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
product=[]

for a,b in df.groupby('TransactionNo').Items:              
    r=[]
    for i in b.values:
        r.append(str(i))
        
    product.append(r)    
```
